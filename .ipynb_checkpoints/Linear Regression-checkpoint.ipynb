# Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import copy, math 
```

```python
#training dataset
X_train = np.array([[2104, 5, 1, 45], [1416, 3, 2, 40], [852, 2, 1, 35]])
y_train = np.array([460, 232, 178])
```

```python
#compute cost function

def computeCost(X, y, w, b):
    m, n = X.shape
    cost = 0.0
    for i in range(m):
            err = np.sum(X[i]*w)+b
            cost = cost + (err-y_train[i])**2
    cost = cost/(2*m)
    return cost
    
```

```python
#Cost function test

b_init = 785.1811367994083
w_init = np.array([ 0.39133535, 18.75376741, -53.36032453, -26.42131618])


computeCost(X_train, y_train, w_init, b_init)
```

```python
#Compute gradient of the features

def computeGradient(X, y, w, b):
    m,n = X.shape
    err = 0
    dj_dw = np.zeros(n)
    dj_db = 0
    for i in range(m):
        err = (np.sum(X[i]*w)+b)-y[i]
        dj_dw = dj_dw + (err*X[i])
        dj_db += err
    dj_dw = dj_dw/m
    dj_db = dj_db/m
    
    return dj_dw, dj_db
        
        
```

```python
temp_dj_dw, temp_dj_db = computeGradient(X_train, y_train, w_init, b_init)
print(f'dj_db at initial w,b: {temp_dj_db}')
print(f'dj_dw at initial w,b: {temp_dj_dw}')
```

```python
#compute gradient descent

def gradientDescent(X, y, w_in, b_in, computeGradient, computeCost, alpha, iters):
    j_history = []
    w_history = []
    b_history = []
    
    w = copy.deepcopy(w_in)
    b = b_in
    
    w_history.append(w)
    b_history.append(b)
    
    for i in range(iters):
        dj_dw, dj_db = computeGradient(X,y,w,b)
        
        w = w - alpha* dj_dw
        b = b - alpha * dj_db
        
        w_history.append(w)
        b_history.append(b)
        
        if i < 100000:
            j_history.append(computeCost(X,y,w,b))
        
        if i% math.ceil(iters/10) == 0: 
            print(f"iterations{i:4d} : cost: {j_history[-1]:8.2f} ")
        
    return w,b,j_history, w_history, b_history
        
```

```python
#Z-score normalization

def Zscore_normalization(X):
    mu = np.mean(X, axis=0)
    sigma = np.std(X, axis=0)
    
    X_norm = (X-mu)/sigma
    
    print(f"X normalized:{X_norm},/n mean: {mu}, /n standard deviation : {sigma}")
    
    return (X_norm, mu, sigma)
    
```

```python
# initialize parameters
initial_w = np.zeros_like(w_init)
initial_b = 0.
iterations = 1000
alpha = 5.0e-7

w_final, b_final, J_hist, W_hist, B_hist = gradientDescent(X_train, y_train, initial_w, initial_b,
                                                    computeGradient, computeCost, 
                                                    alpha, iterations)
print(f"b,w found by gradient descent: {b_final:0.2f},{w_final} ")
m,_ = X_train.shape
for i in range(m):
    print(f"prediction: {np.dot(X_train[i], w_final) + b_final:0.2f}, target value: {y_train[i]}")
```

```python
# initialize parameters
initial_w = np.zeros_like(w_init)
initial_b = 0.
iterations = 1000
alpha = 5.0e-7
X_norm, mu, sigma = Zscore_normalization(X_train)
print(X_norm)
w_norm, b_norm, Jx_hist, Wx_hist, Bx_hist = gradientDescent(X_norm, y_train, initial_w, initial_b,
                                                    computeGradient, computeCost, 
                                                    alpha, iterations)

print(f"b,w found by gradient descent: {b_norm:0.2f},{w_norm} ")
m,_ = X_train.shape
for i in range(m):
    print(f"prediction: {np.dot(X_norm[i], w_norm) + b_norm:0.2f}, target value: {y_train[i]}")
```
